```python
#I have implemented the covariance operation. It will receive two arrays as parameters and will return the covariance(x,y) value. This might help in differentiating dependancies.
def covariance (X, Y): 
    xhat=np.mean(X) 
    yhat=np.mean(Y) 
    epsilon=0
    for x,y in zip (X,Y):
        epsilon=epsilon+(x-xhat)*(y-yhat) 
    return epsilon/(len(X)-1)
```

```python
#Comparision of implemented function and the NumPy function
print (covariance ([2,4,8], [3,6,9]))

print (np.cov([2,4,8], [3,6,9]))

#I have taken the above values as test cases as one of the parameters is petal and sepal(2* criteria so raised the power by 2), other is the 3 types of iris (used the 3 multiples to test)
```

```python
def correlation (X, Y):

    return(covariance(X,Y)/(np.std(X,    ddof=1)*np.std(Y,   ddof=1))) #ddof=1 is unbiased std
```

```python
def correlation_fit(x, y):

    beta = correlation(x, y) * np.std(y, ddof=1) / np.std(x,ddof=1) 
    alpha = np.mean(y) - beta * np.mean(x)

    return alpha, beta
```
